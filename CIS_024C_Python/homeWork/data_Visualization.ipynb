```python
#line chart
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,8,27,64,125]

plt.plot(x,y1,label='first line', color = 'red')
plt.plot(x,y2, label='second line', color = 'blue')

plt.title('my first plot')
plt.xlabel('indices')
plt.ylabel('response values')

plt.legend()

plt.show()
```

```python
#histogram

import numpy as np
import matplotlib.pyplot as plt

mu = 172
sigma = 29

#get weights
weights = np.random.normal(mu, sigma, 1000)

#plot histogram of weights
plt.hist(weights)
plt.title("Distribution of weights of adult males in the US")
plt.show()
```

```python
#pie chart

import matplotlib.pyplot as plt

location_count = [4,7,8,12,3]
location = ['San Jose', 'Santa Clara', "SF", 'Cupertino', 'Fremont']

plt.pie(location_count,
        labels = location,
        autopct = '%1.1f%%',
        startangle = 90,
        explode = (0.4,0,0,0,0.2))

plt.show()
```

```python
# Source: https://www.youtube.com/watch?v=Oh2Dkkswy30

import matplotlib.pyplot as plt

days = [1,2,3,4,5]

sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

plt.plot([],[],color='m', label='Sleeping',linewidth=5)
plt.plot([],[],color='c',label='Eating',linewidth=5)
plt.plot([],[],color='r', label='Working', linewidth=5)
plt.plot([],[],color='k', label='Playing',linewidth=5)

plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Stackplot Example')
plt.legend()
plt.show()
```
